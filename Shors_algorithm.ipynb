```python
import qiskit
import matplotlib
from qiskit import QuantumCircuit
from qiskit.visualization import plot_histogram
import numpy as np
```

```python
def qft_dagger(n):
    qc = QuantumCircuit(n)

    for qubit in range(n // 2):
        qc.swap(qubit, n - qubit - 1)
   
    for j in range(n):
        for m in range(j):
            qc.cp(-np.pi / float(2 ** (j - m)), m, j)
        qc.h(j)
    qc.name = "QFT Dagger"
    return qc
```

```python
def c_amod15(a, power):
    from qiskit import QuantumCircuit
    if a not in [2,7,8,11,13]:
        raise ValueError("'a' must be 2,7,8,11 or 13")
    
    U = QuantumCircuit(4)
    for iteration in range(power):
        if a in [2,13]:
            U.swap(0,1)
            U.swap(1,2)
            U.swap(2,3)
        if a in [7,8]:
            U.swap(2,3)
            U.swap(1,2)
            U.swap(0,1)
        if a == 11:
            U.swap(1,3)
            U.swap(0,2)
        if a in [7,11,13]:
            for q in range(4):
                U.x(q)
    U = U.to_gate()
    U.name = f"{a}^{power} mod 15"
    c_U = U.control()
    return c_U
```

```python
n_count = 8
a = 7
```

```python
qc = QuantumCircuit(n_count + 4, n_count)

for q in range(n_count):
    qc.h(q)

qc.x(n_count + 3)

for q in range(n_count):
    qc.append(c_amod15(a, 2**q), [q] + [i + n_count for i in range(4)])


qc.append(qft_dagger(n_count), qc.qubits[:n_count])
```

```python
qc.measure(range(n_count), range(n_count))
qc.draw()
```
